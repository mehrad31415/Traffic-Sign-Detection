```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
def load(path):
    img=cv.imread(path)
    #opencv reads the image in BGR, thus we have to turn it to RGB
    img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    return img
```

```python
def display(img1,cmap="gray"):
    fig=plt.figure(figsize=(12,18))
    ax=fig.add_subplot()
    ax.imshow(img1,cmap="gray")
```

```python
path="/Users/mehradhq/Downloads/drive-download-20220511T112708Z-001/22.jpeg"
img=load(path)
print ("this is our initial image")
display(img)
```

```python
gray_img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
img_blur=cv.medianBlur(gray_img,9)
```

```python
#adaptive thresholding
thresh=cv.adaptiveThreshold(img_blur, 255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,5,8)
thresh=thresh[:]-255
display(thresh)
```

```python
#canny
med_val=np.median(img_blur)
lower=int(max(0,0.7*med_val))
upper=int(min(255,1.3*med_val))
edges=cv.Canny(img_blur,lower,upper)
display(edges)
```
