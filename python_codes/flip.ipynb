```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
def load(path):
    img=cv.imread(path)
    img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    return img
```

```python
def display(img1,img2,img3,img4):
    fig=plt.figure(figsize=(12,18))
    ax=fig.add_subplot(221)
    ax.imshow(img1)
    ax=fig.add_subplot(222)
    ax.imshow(img2)
    ax=fig.add_subplot(223)
    ax.imshow(img3)
    ax=fig.add_subplot(224)
    ax.imshow(img4)
```

```python
#original image
img=load("/Users/mehradhq/Computer_Vision/Research_2/dataset/train/Prohibition_Signs/17.jpeg")
#flip over the origin
img_origin=cv.flip(img,0)
#flip over the y-axis
img_y=cv.flip(img,1)
#flip over the x-axis
img_x=cv.flip(img,-1)
display(img,img_origin,img_y,img_x)
```
